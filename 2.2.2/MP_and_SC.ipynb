```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
```

###    The Marcenko-Pastur law (Theorem 2.3)
Generate a (Gaussian) random matrix $X$ of dimension $p \times n$.

```python
coeff = 5
p = 100*coeff
n = 10000*coeff
c = p/n

X = np.array(np.random.randn(p,n))
```

Plot the empirical eigenvalues of $\hat C = \frac1n X X^T$ versus the (limiting) Marcenko-Pastur law.

```python
SCM = X@X.T/n
a = (1-np.sqrt(c))**2
b = (1+np.sqrt(c))**2
edges=np.linspace(a,b,60);

# Limiting measure
mu=np.sqrt( (edges-a)*(b-edges) )/2/pi/c

# Empirical histogram

plt.hist(np.linalg.eig(SCM)[0],bins=edges,weights=1/p/(edges[1]-edges[0])*np.ones(p),label='Empirical eigenvalues')
plt.plot(edges,mu/edges,'r',label='Marcenko-Pastur law')
_=plt.legend()
```

### The Wigner semi-circle lawï¼ˆTheorem 2.4) 

Generate a (Gaussian) **symmetric** random matrix $X$ of size $n \times n$.

```python
coeff = 5
n=200*coeff

Z=np.array(np.random.randn(n,n))
Z_U = np.triu(Z)
X = np.triu(Z) + np.triu(Z,1).T
```

Plot the empirical eigenvalues of $\frac{X}{\sqrt{n}}$ versus the limiting semicircle law.

```python
edges=np.linspace(-2,2,50)

mu = np.sqrt( np.maximum.reduce( [4 - edges**2, np.zeros(len(edges))] ) )/2/pi

plt.hist(np.linalg.eig(X/np.sqrt(n))[0],bins=edges,weights=1/n/(edges[1]-edges[0])*np.ones(n),label='Empirical eigenvalues')
plt.plot(edges,mu,'r',label='Wigner semi-circle law')
_ = plt.legend()
```
